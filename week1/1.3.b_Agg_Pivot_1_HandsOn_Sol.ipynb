## Aggregation of data and pivot tables - Hands-on solution
---

### <span style="color:red">Hands-on exercise</span>

* Create the MultiIndex groupby object from above with `pivot_table()`
* Create a pivot table that shows the data for `Quantity` as a sum *and* as an average value by month and location

```python
import pandas as pd
```

Import the data as CSV:

```python
sales = pd.read_csv('./data/pivtab_en.csv', sep=';',parse_dates=True)
```

```python
sales.head()
```

```python
sales.pivot_table('Quantity', index=['MainProductGroup', 'Region'], 
                   aggfunc='sum')
```

```python
sales.pivot_table('Quantity', index=['Month'], columns=['Region'],
                   aggfunc=['sum','mean'])
```

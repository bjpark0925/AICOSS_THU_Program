## Principles of good visualizations - IBCS Standards
---

> All images below are distributed by the [IBCS Association](https://www.ibcs.com) as part of the IBCS standard (CC BY-SA).

Business communication complies with the [IBCS standards](https://www.ibcs.com/de/standards/#%3F=) (**I**nternational **B**usiness **C**ommunication **S**tandards) if it fulfills the rules of the seven areas that form the acronym "SUCCESS":

* **S**AY - *Message conveyed*
* **U**NIFY - *Use semantic notation*
* **C**ONDENSE - *Increase information density*
* **C**HECK - *ensure visual integrity*
* **E**XPRESS - *Select suitable visualization*
* **S**IMPLIFY - *Avoid the superfluous*
* **S**TRUCTURE - *Structure content*

The seven rule areas of the SUCCESS formula can be assigned to **three categories**, which form the following three chapters of the IBCS standards. The contents of these chapters often go far beyond the design of diagrams and charts and shed light on all aspects of business communication. Since a complete presentation of these aspects goes beyond the scope of this teaching unit, we will deal with *some points* below that can have a direct impact on the design of visualizations.

### Conceptual Rules

Conceptual rules help to convey content clearly using a suitable **storyline**. They comprise the first part of the IBCS standards with the SUCCESS rule sets SAY and STRUCTURE.

#### Say - Convey a message

SAY encompasses all aspects of **message transmission** to the recipients of reports and presentations. Conveying messages means that reports and presentations, both as a whole and in their individual components, want to communicate something to the recipients.

Perceptual rules help to convey content clearly through a suitable visual design. They form the second part of the IBCS standards with the SUCCESS rule sets EXPRESS, CHECK, CONDENSE and SIMPLIFY.

### Semantic Rules

Semantic rules help to clearly convey content using a standardized notation (IBCS notation). Together with the SUCCESS rules UNIFY, they form the third part of the IBCS standards.

#### Unify - Apply semantic notation

This topic covers all aspects of **semantic notation** in reports and presentations. Applying semantic notation means that reports and presentations follow this guiding principle: Similar content should be visualized in a similar way; what looks the same should mean the same. Conversely, what is not the same should not look the same.

### Story > Rules

Despite all the rules, it is easy to lose sight of the goal of visualizations: to tell a moving *data story*. The visualizations should be a tool to make the underlying meaning of the data accessible and exciting for the recipients of the story. Of course, the facts should not be distorted with "visual tricks" (section [Check - Ensure visual integrity](#Check---Ensure-visual-integrity)).

Sometimes creative visualization methods are needed to tell a data story. [Here](https://mkorostoff.github.io/1-pixel-wealth/), for example, the wealth of Jeff Bezos is compared using a "bar chart". ## Matplotlib
---

![matplotlib1](img/vis_matplotlib1.png)

[Matplotlib](https://matplotlib.org/) is the standard library for creating plots and visualizations in Python. It is one of the first large plotting libraries in Python. The API is relatively "low-level" - this means that plots can be configured down to the smallest detail. The disadvantage of this is certainly that a few lines of code are often necessary to achieve the goal.  

Matplotlib uses two different [interfaces](https://matplotlib.org/stable/users/explain/figure/api_interfaces.html) to create plots: **Object-oriented** and **Pyplot**. Furthermore, Matplotlib visualizes the data as a [Figure](https://matplotlib.org/stable/api/figure_api.html?highlight=figure#matplotlib.figure.Figure), which in turn can contain one or more [Axes](https://matplotlib.org/stable/api/axes_api.html). We will take a closer look at these terms below.

**Resources**:

* [Pyplot tutorial](https://matplotlib.org/stable/tutorials/introductory/pyplot.html)
* [Pyplot functions](https://matplotlib.org/stable/api/pyplot_summary.html#plotting-commands)
* [Format strings](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html)
* Matplotlib [Cheatsheet](https://s3.amazonaws.com/assets.datacamp.com/blog_assets/Python_Matplotlib_Cheat_Sheet.pdf)
* Chart [types](https://matplotlib.org/stable/gallery/index.html)
* [Artists](https://matplotlib.org/stable/api/artist_api.html) (objects rendered on a figure)

### Figure

The figure is the superordinate visualization element in Matplotlib. The individual plots in terms of axes are managed by the figure. A figure can contain any number of axes (at least one).

![Figure](img/vis_matplotlib2.png)

**Note:** All objects that are rendered by Matplotlib on a figure are referred to as [Artists](https://matplotlib.org/stable/api/artist_api.html). This would be the entry point for accessing individual objects.

### Axes

This is what is meant by a "plot", it is the area of the image with the data space. A figure can contain many axes, but an
Axes object can only be in one figure. Axes contains two (or three in the case of 3D) Axis objects (note the *difference between Axes and Axis*!) that take care of the data boundaries (the data boundaries can also be controlled by the methods `axes.Axes.set_xlim()` and `axes.Axes.set_ylim()`). Each axes has a title (set via `set_title()`), an x-label (set via `set_xlabel()`) and a y-label (set via `set_ylabel()`).

Axes and their functions are the primary entry point for working with the object-oriented API.

### Pyplot API

Each time we call a [Pyplot function](https://matplotlib.org/stable/api/pyplot_summary.html#plotting-commands), we can use functions to make a change to a figure. When using them, we usually do not need to concern ourselves with the separation of figures and axes. If we apply different functions to the `plt` object, then states are preserved across these function calls. This approach is also called `stateful`.

Let's take a look at the Pyplot API using an example:

````python
import matplotlib.pyplot as plt # same as: from matplotlib import pyplot as plt
````

````python
import numpy as np
import pandas as pd
````

````python
df = pd.read_csv('./data/imdb_ratings.csv')
````

````python
df.head()
````

Transformation of ratings into categories:

````python
df['Rating_cat'] = pd.cut(df['Rating'], bins=3, labels=['Low', 'Medium', 'High'])
````

````python
df_pivot = df.pivot_table(index=['Year'], values='Revenue (Millions)', aggfunc='sum')
````

````python
df_pivot
````

Matplotlib recommends using numpy arrays as input. We can extract these from the DataFrame with `.values`:

````python
x = df_pivot.index.values
y = df_pivot['Revenue (Millions)'].values
````

````python
plt.figure(figsize=(8,5), dpi=100)
plt.plot(x,y, color='k', linestyle='--', marker='o')
plt.xlabel('Year')
plt.ylabel('Revenue (Millions)')
plt.title('Movies Revenue per Year')
plt.text(2010.7, 4800, 'Outlier!', bbox=dict(facecolor='red', alpha=0.5))
plt.tight_layout() # Adds padding and sometimes prevents cropping
plt.show()
````

#### Connection to pandas

The `.plot` method of Pandas is a [wrapper](https://realpython.com/primer-on-python-decorators/) for `plt.plot`. This means we can manipulate the plot with the [Pyplot functions](https://matplotlib.org/stable/api/pyplot_summary.html#plotting-commands):

````python
plt.figure(figsize=(8,5), dpi=100)
df_pivot.plot(kind='line', legend=False, ax=plt.gca())
plt.ylabel('Revenue (Millions)')
plt.title('Movies Revenue per Year')
plt.text(2010.7, 4800, 'Outlier!', bbox=dict(facecolor='red', alpha=0.5))
plt.text(2006,10000, r'Even Tex is possible: $\sigma \beta \omega \mu \infty$')
plt.show()
````

### Object-oriented API

With the object-oriented API, we define the **Figures and Axes** explicitly:

````python
fig, ax = plt.subplots(figsize=(8,5), dpi=100)

ax.plot(x,y, color='k', linestyle='--', marker='o')
ax.set_xlabel('Year')
ax.set_ylabel('Revenue (Millions)')
ax.set_title('Movies Revenue per Year')
ax.text(2010.7, 4800, 'Outlier!', bbox=dict(facecolor='red', alpha=0.5))
#plt.savefig('plot2.png')
plt.show()
````

This allows us to insert new axes into the existing figure and thus overlay plots, for example:

````python
fig, ax = plt.subplots(figsize=(8,5), dpi=100)

ax.plot(x,y, color='k', linestyle='--', marker='o')
ax.set_xlabel('Year')
ax.set_ylabel('Revenue (Millions)')
ax.set_title('Movies Revenue per Year')
ax.text(2010.7, 4800, 'Outlier!', bbox=dict(facecolor='red', alpha=0.5))

ax2 = fig.add_axes([0.25, 0.55, 0.3, 0.2]) # [left, bottom, width, height]
ax2.set_xlabel('Year')
ax2.set_ylabel('Revenue (Millions)')
ax2.set_title('Movies Revenue per Year')
ax2.plot(x,y, color='b', linestyle='--', marker='x')

plt.show()
````

Let us illustrate the adaptability of Matplotlib with this bar chart according to **IBCS standard**:

**Matplotlib Standard**

````python
fig, ax = plt.subplots(figsize=(7,4), dpi=90)
ax.bar(x,y)
plt.show()
````

**IBCS**

````python
fig, ax = plt.subplots(figsize=(8,5), dpi=100)

ax.bar(x,y, color='grey')
ax.set_title('Movies Revenue in Millions\n from 2006 to 2016')
ax.set_frame_on(False) # remove axis
ax.set_xticks(x) # show all labels
ax.tick_params(bottom=False,labelbottom=True, labelleft=False, left=False) # Ticks und Ticklabels entfernen

# annotate bars with values in x
for x_val, y_val in zip(x,y):
    ax.text(x_val, y_val+100, str(int(y_val)), color='k', horizontalalignment="center")
plt.show()
````

**Conclusion on the APIs:** The Pyplot API is less flexible than the object-oriented API and is primarily suitable for simpler plots. However, since the changeover from Pyplot to object-oriented requires only a few steps, you may want to consider using the object-oriented API right away.

### Simple plots

````python
import seaborn as sns
````

With [sns.set_theme()](https://seaborn.pydata.org/generated/seaborn.set_theme.html#seaborn.set_theme) we set the plotting style from *Matplotlib-Default* to *Seaborn*:

````python
sns.set_theme()
````

````python
plt.figure(figsize=(8,5), dpi=100)
sns.barplot(data=df_pivot.reset_index(), x='Year', y='Revenue (Millions)');
````

A **Pair Plot** with categorical variable `Rating_cat` colored:

````python
sns.pairplot(df, hue='Rating_cat');
````

### Color palette

**Ressourcen**

* [Seaborn Color Palettes](https://seaborn.pydata.org/tutorial/color_palettes.html)
* [Matplotlib Colormaps](https://matplotlib.org/stable/tutorials/colors/colormaps.html)

Seaborn uses a [color palette](https://seaborn.pydata.org/tutorial/color_palettes.html) for most plots (unless otherwise specified). The following palette is stored by default:

````python
sns.color_palette()
````

We can customize the color palette by selecting from the large number of ready-made color palettes, for example:

````python
plt.figure(figsize=(8,5), dpi=100)
sns.barplot(data=df_pivot.reset_index(), x='Year', hue='Year', y='Revenue (Millions)', palette='gray', legend=False);
````

We can also set the color palette for all plots with the following command:

````python
sns.set_palette('tab10') # default palette
````

**Note:** For most plots, this will only adjust the color value for variables in the `hue` parameter!

We can also use our own color palettes:

````python
colors = ["#FC631E", "#DADADA", '#097FAC']
````

````python
sns.color_palette(colors)
````

````python
my_palette = sns.light_palette(colors[0], n_colors=11, as_cmap=False)
my_palette
````

````python
plt.figure(figsize=(8,5), dpi=100)
sns.barplot(data=df_pivot.reset_index(), x='Year', hue='Year', y='Revenue (Millions)', palette=my_palette, legend=False);
````

Seaborn provides some [widgets](https://seaborn.pydata.org/api.html#palette-widgets) to select palettes directly in Jupyter:

````python
sns.choose_light_palette()
````

### Interaction with Matplotlib

Since Seaborn is based on Matplotlib, we can make use of the Matplotlib structures `Figure` and `Axes` to manipulate plots. For example, Seaborn plots can be restricted to corresponding axes in subplots:

````python
f, axs = plt.subplots(1, 2, figsize=(8, 4), dpi=100, gridspec_kw=dict(width_ratios=[4, 3]))
sns.scatterplot(data=df, x="Runtime (Minutes)", y="Revenue (Millions)", hue="Rating_cat", ax=axs[0])
sns.histplot(data=df, x="Rating_cat", hue="Rating_cat", shrink=.8, alpha=.8, legend=False, ax=axs[1])
f.tight_layout()
````
